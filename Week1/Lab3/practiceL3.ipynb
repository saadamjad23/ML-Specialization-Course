```python
# Importing Libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])
w = 200
b = 100
```

```python
# cost function
def compute_cost(x, y, w, b):
    cost_sum = 0
    m = x.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb = w * x[i] + b
        cost = (f_wb - y[i])**2
        cost_sum = cost_sum + cost
    total_cost = (cost_sum)/2 * m
    
    return total_cost

final_cost = compute_cost(x_train, y_train, w, b)
final_cost
```

```python
# ploting graph
plt.scatter(x_train, y_train, marker = 'X', c = 'r')
plt.plot(x_train, y_train, c = 'b')
plt.title("House Price Prediction")
plt.xlabel("Size(1000sq feet)")
plt.ylabel("Price($1000)")
plt.show()
```

```python
x_train2 = np.array([1.0, 1.7, 2.0, 2.5, 3.0, 3.2])
y_train2 = np.array([250, 300, 480,  430,   630, 730,])
```

```python
compute_cost(x_train2, y_train2, w, b)
```

```python
# ploting graph
plt.scatter(x_train2, y_train2, marker = 'X', c = 'r')
plt.plot(x_train2, y_train2, c = 'b')
plt.title("House Price Prediction")
plt.xlabel("Size(1000sq feet)")
plt.ylabel("Price($1000)")
plt.show()
```

```python
def compute_model(x, w, b):
    m = x.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x[i] + b
    return f_wb
```

```python
model = compute_model(x_train2, w, b)
print(f"Prediction : {model}")
```

```python
# ploting graph
plt.scatter(x_train2, y_train2, marker = 'X', c = 'r')
plt.plot(x_train2, model, c = 'b')
plt.title("House Price Prediction")
plt.xlabel("Size(1000sq feet)")
plt.ylabel("Price($1000)")
plt.show()
```
