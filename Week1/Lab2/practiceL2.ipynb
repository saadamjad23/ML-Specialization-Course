```python
#importing Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# intializing values of x_train & y_train
x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])
print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
```

```python
# defining m (number of examples)
print(x_train.shape)
m1 = x_train.shape[0]   # tells number of rows in an array
print(f"m = {m1}")

# Second Method
m2 = len(x_train)
print(f"m = {m2}")
```

```python
# accessing ith training example
for i in range(2):
    x_i = x_train[i]
    y_i = y_train[i]
    print(f"x_{i} = {x_train[i]}")
    print(f"y_{i} = {y_train[i]}")
```

```python
# plotting the data
plt.scatter(x_train, y_train, marker = 'X', c = 'r')
plt.title("Data Points")
plt.xlabel("Size of House in 1000 sqrt")
plt.ylabel("Price of House in $1000")
plt.show()
```

```python
# initializing values of w & b
w = 100
b = 100
print(f"w = {w}")
print(f"b = {b}")
```

```python
# compute model outputs
def compute_model(x, w, b):
    m = x.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x_train[i] + b
    return f_wb

print(compute_model(x_train, w, b))
```

```python
temp_f_wb = compute_model(x_train, w, b)

# plotting the data
plt.scatter(x_train, y_train, marker = 'X', c = 'r', label = "Actual Values")
plt.plot(x_train, temp_f_wb, c= 'b', label = "Our Prediction")
plt.title("Data Points")
plt.xlabel("Size of House in 1000 sqrt")
plt.ylabel("Price of House in $1000")
plt.show()
```
